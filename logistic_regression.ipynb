```
import numpy as np
import matplotlib.pyplot as plt
```

```
def gradient_descent(x,y,theta,alpha,m,num_iterations):
  xtrans=x.transpose()
  
  cost_ar=np.zeros(num_iterations)
  
  for i in range(num_iterations):
    
    h= (1.0/(1+np.exp(np.dot(x,theta))))
    
    
    cost =-y.dot(np.log(h))-(1-y).dot(np.log(h))
    print("cost", cost)
    
    hypothesis=x.dot(theta)
    
    loss = hypothesis -y
    
    cost= np.sum(loss**2/2*m)
    
    cost_ar[i]=cost
    
    #print("num interations",i,"gradinet cost=",cost)
    
    gradient= np.dot(xtrans ,loss)
    
    theta=theta-alpha*gradient
    
  #______  graph of cost vs number of iterations at the end of the cost list
  
  
  
  fig, ax = plt.subplots()  
  ax.plot(np.arange(num_iterations), cost_ar, 'r')  
  ax.set_xlabel('Iterations')  
  ax.set_ylabel('Cost')  
  ax.set_title('Error vs. Training Epoch')
    
  return(theta)
    
  
```

```
numpoints =10
x=np.zeros(shape = (numpoints,2))
y=np.zeros(shape = numpoints)
theta=np.ones(2)
  
for i in range(0 ,numpoints):
    
  x[i][0]=1
  x[i][1]=i
  
  
for i in range(5,10):
  y[i]=1
  
print(x)
print(y)

# alpha=0.0001
# m= number of datapoint in training data set (10)

theta=gradient_descent(x,y,theta,0.0001,10,5000)
print(theta)
```

```
z=x.T[1]


def plot(x,y,b):
  plt.scatter(x, y, color = "m",
               marker = "o", s = 30)
    # predicted response vector
  y_pred = b[0] + b[1]*x
    # plotting the regression line
  plt.plot(x, y_pred, color = "r")
  plt.plot(x,y,color ="g")
  plt.xlabel("x")
  plt.ylabel("y")
  
  plt.show()
```

```
plot(z,y,theta)
```
