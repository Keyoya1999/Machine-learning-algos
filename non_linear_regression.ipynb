```
import numpy as np
import random
import matplotlib.pyplot as plt
```

```
def gradient_descent(x,y,theta,alpha,m,num_iterations):
  
  xtans=x.transpose()
  
  cost_ar=np.zeros(num_iterations)
  
  for i in range(num_iterations):
    
    hypothesis=np.dot(x.T, theta)
    
    
    loss= hypothesis - y
    
    #print(loss)
    #print(loss**2)
    #break
    
    cost= np.sum(loss**2/2*m)
    cost_ar[i]=cost
    
    print("num interations",i,"    cost=",cost)
    
    gradient= np.dot(x ,loss)
    
    theta=theta-alpha*gradient
    
    #basic mistake of simultanous update of theta variable
  #fig, ax = plt.subplots()  
  #ax.plot(np.arange(num_iterations), cost_ar, 'r')  
  #ax.set_xlabel('Iterations')  
  #ax.set_ylabel('Cost')  
  #ax.set_title('Error vs. Training Epoch')  
  return(theta)
```

```
def plot(x,y,b):
  plt.scatter(x, y, color = "m",
               marker = "o", s = 30)
    # predicted response vector
  y_pred = b[1] + b[0]*x +b[2]*x*x
    # plotting the regression line
  plt.plot(x, y_pred, color = "g")
  plt.plot(x,y,color ="g")
  plt.xlabel("x")
  plt.ylabel("y")
  
  plt.show()

  
```
