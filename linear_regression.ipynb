```
import numpy as np
import random
import matplotlib.pyplot as plt
```

```
def gradient_descent(x,y,theta,alpha,m,num_iterations):
  
  xtans=x.transpose()
  
  cost_ar=np.zeros(num_iterations)
  
  for i in range(num_iterations):
    
    hypothesis=np.dot(x, theta)
    
    
    loss= hypothesis - y
    
   
    cost= np.sum(loss**2/2*m)
    cost_ar[i]=cost
    
    print("num interations",i,"    cost=",cost)
    
    gradient= np.dot(xtans ,loss)
    
    theta=theta-alpha*gradient
    
    
    
    
  # ---- ploting cost vs num of iterations at the end of the list
  
  
  
  fig, ax = plt.subplots()  
  ax.plot(np.arange(num_iterations), cost_ar, 'r')  
  ax.set_xlabel('Iterations')  
  ax.set_ylabel('Cost')  
  ax.set_title('Error vs. Training Epoch')
  
  return(theta)
    
    
  
 
```

```
def gendata(numpoints ,bias ,variance):
  x=np.zeros(shape = (numpoints,2))
  y=np.zeros(shape = numpoints)
  
  for i in range(0 ,numpoints):
    
    x[i][0]=1
    x[i][1]=i
    
    y[i] = (i+bias) +random.uniform(0,1)*variance
    
  return (x,y)
```

```
x, y = gendata(10, 10, 5)
m, n = np.shape(x)
numIterations= 20000
alpha = 0.0002
theta = np.ones(n)
print(x)
theta = gradient_descent(x, y, theta, alpha, m, numIterations)
print(theta)
```

```

z=x.T[1]


def plot(x,y,b):
  plt.scatter(x, y, color = "m",
               marker = "o", s = 30)
    # predicted response vector
  y_pred = b[0] + b[1]*x
    # plotting the regression line
  plt.plot(x, y_pred, color = "r")
  plt.plot(x,y,color ="g")
  plt.xlabel("x")
  plt.ylabel("y")
  
  plt.show()
```

```
plot(z,y,theta)
```
