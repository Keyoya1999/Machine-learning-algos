# **Support Vector machine implementation using linear and poly functional kernels**

```
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets,svm
import sklearn
```

```

X,y=sklearn.datasets.make_blobs(n_samples=5000, n_features=2, centers=3, cluster_std=0.8)
```

```
C = 1.0 # SVM regularization parameter

svc = svm.SVC(kernel='linear',C=1,gamma='auto').fit(X,y)
svc_poly =svm.SVC(kernel='poly',degree=5,C=1).fit(X,y)
svc_rbf=svm.SVC(kernel='rbf',C=1).fit(X,y)


print(svc)

plt.scatter(X[:,0],X[:,1])
#ploting the intial data set
```

```
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1            #creating mesh grids for contour plots
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
h = (x_max / x_min)/100
x_mesh=np.arange(x_min, x_max, -h)
y_mesh=np.arange(y_min, y_max, -h)
xx, yy = np.meshgrid(x_mesh,y_mesh)


```

**Ploting the classifier using  Support vetor machine using linear kernels**


```
plt.subplot(1, 1, 1)                                   #ploting the contour plots and the data points
Z = svc.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)

plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.viridis)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xlim(xx.min(), xx.max())
plt.title('SVC with linear kernel')
plt.show()
```

**Ploting the classifier using  Support vetor machine using polynomial function of degree 3 kernels**

```
plt.subplot(1, 1, 1)
Z = svc_poly.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)

plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.viridis)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xlim(xx.min(), xx.max())
plt.title('SVC with poly functional kernel with degree 3')
plt.show()
```

**Ploting the classifier using  Support vetor machine using rbf kernels**

```
plt.subplot(1, 1, 1)
Z = svc_rbf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)

plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.viridis)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xlim(xx.min(), xx.max())
plt.title('SVC with rbf kernel')
plt.show()
```
